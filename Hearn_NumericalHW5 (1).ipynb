```python
# Problem 2 - Power Method
import numpy as np

A = np.array([[5,4,1,1], [4,5,1,1], [1,1,4,2], [1,1,2,4]])
A1 = np.linalg.inv(A)
n = len(A)
x = np.random.rand(A.shape[0])
M = 100

def Phi(x):
    phix = x.item(1)
    return phix

def PowerMethod(A, x, M): #normalized
    for _ in range(M):
        y = np.dot(A, x)
        r = Phi(y)/Phi(x)
        y_norm = np.linalg.norm(y)
        x = y/y_norm
    return x,r

def PowerUnNorm(A, x, M): #un-normalized
    for _ in range(M):
        y = np.dot(A, x)
        r = Phi(y)/Phi(x)
        x = y
    return x,r

PMN = PowerMethod(A,x,M)
PMI = PowerMethod(A1,x,M) #inverse PM

print PMN
print PMI

PUN = PowerUnNorm(A,x,M)
print PUN

# Notes - the un-normalized Power Method is giving nowhere near true results.
```

```python
# Problem 3 - Gram-Schmidt, traditional and modified
import numpy as np

test = np.random.rand(20,10)
test2 = np.fromfunction(lambda i, j: ((2*i-21)/19)**(j-1), (20, 10))
n = len(test)
m = len(np.transpose(test))
    
def GramSchmidt(m,n,A):
    C = np.zeros((n,m))
    Q = np.zeros((n,m))
    R = np.zeros((n,m))
    for j in xrange(m):
        for i in xrange(j):
            R[i,j] = np.dot(A[:,j],Q[:,i])
        C[:,j] = A[:,j]
        for i in xrange(j):
            C[:,j]=C[:,j]-R[i,j]*Q[:,i]
        R[j,j] = np.linalg.norm(C[:,j])
        Q[:,j] = C[:,j]/R[j,j]
    return R, Q

def GramSchmidtMod(m,n,A):
    for k in xrange(m):
        A[:,k] = A[:,k]/np.linalg.norm(A[:,k])
        for j in range(k+1,m):
            A[:,j] = A[:,j] - np.dot(A[:,j],A[:,k])*A[:,k]
    return A

A = GramSchmidt(m,n,test)
A2 = GramSchmidt(m,n,test2)
B = GramSchmidtMod(m,n,test)
B2 = GramSchmidtMod(m,n,test2)

print A
print A2
print B
print B2

#note - ran out of time to add testing criterion
```
